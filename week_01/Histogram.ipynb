```python
import pandas as pd
```

```python
lifeexp = pd.read_excel('data/gapminder_lifeexpectancy.xlsx')
```

```python
lifeexp
```

```python
lifeexp.describe()
```

```python
lifeexp = lifeexp[[1950, 1975, 2000, 2015]]
```

```python
import matplotlib.pyplot as plt

lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
lifeexp[1950].hist(bins=5)
lifeexp[1975].hist(bins=10)
lifeexp[2000].hist(bins=20, facecolor='#ff0000',
alpha=0.75,
histtype='bar',)
plt.title('Life Expectancy')
plt.xlabel('Age')
plt.ylabel('Countries')
plt.axis([20.0, 100.0, 0.0, 60.0])
```

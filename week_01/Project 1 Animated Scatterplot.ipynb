# Creating an Anmiated Scatterplot

```python
import pandas as pd

fert = pd.read_csv('data/gapminder_total_fertility.csv', index_col=0)
```

```python
fert
```

```python
life = pd.read_excel('data/gapminder_lifeexpectancy.xlsx',index_col=0)
life
```

```python
fert.shape
```

```python
life.columns
```

```python
fert.columns
```

```python
fert.columns = fert.columns.astype(int)
```

```python
fert.columns
```

```python
fert.index
```

```python
fert.index.name = 'country'
```

```python
fert = fert.reset_index()
```

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
life.index
```

```python
life.index.name = 'country'
```

```python
life = life.reset_index()
```

```python
life = life.melt(id_vars= 'country', var_name ='year', value_name = 'life_expectancy')
```

```python
population = pd.read_excel('data/gapminder_population.xlsx', index_col = 0)
```

```python
population
```

```python
population.shape
```

```python
population.columns
```

```python
population.index
```

```python
population.index.name = 'country'
```

```python
population = population.reset_index()
```

```python
population = population.melt(id_vars='country', var_name='year', value_name='population')
```

```python
df = fert.merge(population)
```

```python
df = df.merge(life)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset = df.loc[df['country'].isin(['France', 'Germany', 'Sweden'])]
sns.lineplot(x='year', y='life_expectancy', hue='country',
            data=df_subset)
```

```python
df_subset
```

```python
sns.scatterplot(x='year', y='life_expectancy', hue='country',
            data=df_subset, alpha=0.6)
```

```python
for i in range(1960,2016):
    df_animation = df_subset.loc[df_subset['year'] == i]
    sns.scatterplot(x='year', y='life_expectancy', hue='country',
            data=df_animation, alpha=0.6)
    plt.axis((1960, 2015, 69, 83))
    plt.savefig(f'scatterplots/life/life{i}.png', dpi=150)
    plt.close()
```

```python
for i in range(1960,2016):
    df_animation = df_subset.loc[df_subset['year'] == i]
    sns.scatterplot(x='year', y='life_expectancy', hue='country', size = 'population', 
            data=df_animation, alpha=0.6)
    plt.axis((1960, 2015, 69, 83))
    plt.savefig(f'scatterplots/population_life/population_life{i}.png', dpi=150)
    plt.close()
```

```python
sns.scatterplot(x='year', y='life_expectancy', hue='country', size = 'population', 
            data=df_subset, alpha=0.6)
plt.axis((1960, 2015, 69, 83))
```

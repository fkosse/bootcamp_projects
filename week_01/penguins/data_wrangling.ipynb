### 2. Load fragmented penguin data
- One ``.csv`` file for each species
- How can we combine them all into one DataFrame?

```python
adelie = pd.read_csv('penguins_adelie.csv')
chinstrap = pd.read_csv('penguins_chinstrap.csv')
gentoo = pd.read_csv('penguins_gentoo.csv')
```

```python
df = pd.concat([adelie, chinstrap, gentoo], axis = 0) #axis = 0 means rows, 1 means columns
df.shape
```

```python
import time
import random
# scrape multiple pages 

eminem = []
links_100 = random.choices(links, k=300)

for l in links_100: #(try a small list frist (3 entries))
    lyrics = []
    lyrics_clean = []
    lyrics_clean_clean = []
    lyrics_clean_clean_clean = []
    new_url = 'https://lyrics.com/lyric/' + l
    response = requests.get(new_url)
    lyrics = re.findall(r'<pre id="lyric-body-text".+<\/pre>', response.text, re.DOTALL)
    try:
        lyrics_clean = re.sub(r'<.+?>','',lyrics[0])
        lyrics_clean_clean = re.sub(r'\n',' ',lyrics_clean)
        lyrics_clean_clean_clean = re.sub(r'\\','',lyrics_clean_clean)
        eminem.append((lyrics_clean_clean_clean, 'eminem'))
        print(new_url, response.status_code)
        time.sleep(10) # do nothing for 10 seconds
    except IndexError:
        print(new_url, response.status_code)
        time.sleep(10) # do nothing for 10 seconds
```

```python
import pandas as pd
eminem_df = pd.DataFrame(eminem, columns = ['lyrics','artist'])
eminem_df
```

```python
eminem_df.to_csv('data/eminem.csv')
```

```python
import seaborn as sns

df = sns.load_dataset("anscombe")
```

```python
df
```

```python
sns.scatterplot(data=df, x= 'x', y = 'y', hue = 'dataset')  
```

```python
df.describe()
```

```python
df.groupby('dataset')[['x','y']].describe()
```

```python
df.groupby('dataset')[['x','y']].agg(['mean', 'std'])
```

```python
sns.pairplot(data=df, hue='dataset')
```

```python
sns.heatmap(df.groupby('dataset').corr(), 
       
            cmap='Blues',
            linewidths= 1,
            annot=True)
```

```python
df.groupby('dataset').plot.scatter(x='x', y='y')
```
